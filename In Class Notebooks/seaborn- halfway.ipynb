### Exploring relationships within our dataset
We want explore the relationship between our numerical columns, there are 2 ways.
<br>
-- 1) Correlation analysis <br>
-- 2) Pairplot

```python
#correlation analysis
plt.figure(figsize=(10,7))

corr = censusdf.corr()

#plot our correlation data
sns.heatmap(corr, annot=True)
```

### Values closer to -1 or +1, implies a strong correlation relation between our features.
Values closer to 0, indicates no correlation!

Based on what we see above, our features have little to no correlation.

```python
#second method
#pairplot analysis

#A pairs plot allows us to see both distribution of single variables and relationships 
#between two variables. Pair plots are a great method to identify trends for follow-up analysis 
#and, fortunately, are easily implemented in Python!
sns.pairplot(censusdf)
```

#### As expected from our correlation analysis seen in method 1, the pairplot method also shows no relationship that can be directly observed, between our numerical features!

### Let's plot more charts to visually inspect our data

As part Exploratory Data Anaysis: We want to find out more insights about our dataset!

As we chart each visual, let's go ahead and obtain at least 1 insight or observation from each chart!

```python
style.use("ggplot")
```

```python
censusdf.info()
```

```python
# chart 1) take a deeper look at capital gain and capital loss
# we want to use a scatterplot to visual them
# we will add on the marital_status 
plt.figure(figsize=(9,7))
sns.scatterplot(x='capital_loss',
                y='capital_gain',
                hue='marital_status',
                data=censusdf)
```

```python
#some observations
#1) there are outliers! on boths sides
#2) the datapoints have concentrations - namely, around 100k for gains 
#   and 2k for loss
#3) married to civ spouse takes up more of our dataset
#4) separated/widowed tends to have higher capital loss
```

```python
#chart 2) we want to add age to our chart
plt.figure(figsize=(9,7))
sns.scatterplot(x='capital_loss',
                y='capital_gain',
                hue='marital_status',
                size='age',
                sizes=(0,300),
                data=censusdf)
```

```python
#1) outliers tend to be older
#2) younger people have more loss
```

```python
#Chart 3) plot age vs hours_per_week, use a line plot
plt.figure(figsize=(10,7))
sns.lineplot(x='age', y='hours_per_week',
             hue='gender',
             data=censusdf)
```

```python
#Some observations
#1) the max work hours per week is done by people whom are age 28-60
#2) above 60, people tend to work lesser hours - could be that they semi-retired
#3) lesser data points for age 80+
#4) 16-28 , there is a steady increase in the number of work hours
#5) Until age 70, females have clocked lesser hours than males 
```

```python
#chart 4) facet grids - "generate" charts - use our chart 1 to generate
plt.figure(figsize=(9,7))
sns.relplot(x='capital_loss',
            y='capital_gain',
            hue='marital_status',
            size='age',
            sizes=(0,300),
            col='gender',
            data=censusdf)
```

```python
#chart 5) try to understand the distribution of age with 
# respect to marital status

#boxplot
plt.figure(figsize=(16,8))
sns.boxplot(x='age', y='marital_status', data=censusdf)
```

```python
#1) married and widowed have long whiskers i.e. very much distributed in age
#2) more younger people are never-married (yet), and more widowed in the older people
#3) some young people have spouse have passed on, outliers before 25 yr
```
