## Some Simple Visualisation

```python
import pandas as pd
```

```python
df = pd.DataFrame({'name': ['Jack', 'John', 'Joe'],
                    'salary': [1500,25000, 125000]})
df
```

```python
#central tendency - mean or average
df['salary'].mean()
```

```python
#central tendency - median
df["salary"].median()
```

```python
newdf = pd.DataFrame({'name': ['Jack', 'John', 'Joe', "Megan", "Nicole"],
                    'salary': [1500,25000, 125000, 7500, 7500]})

newdf
```

```python
#central tendency - mode
newdf["salary"].mode()
```

```python
#box and whisker plot
%matplotlib inline
newdf.boxplot(column='salary')
```

```python
newdf.describe()
```

### The matplotlib library
Base libary

```python
#import matplotlib
import matplotlib.pyplot as plt
```

```python
#line plot
years = [2020, 2021, 2022, 2023, 2024, 2025]
totalPop = [8000000, 12000000, 3000000, 5000000, 3600000, 4600000]

plt.plot(years, totalPop)# default .plot gives us a line chart!
plt.title("Year vs Population on Mars")
plt.xlabel("Years")
plt.ylabel("Total population")
plt.ticklabel_format(style='plain')
```

```python
# scatter plots
# they us to understand relationship
temp = [30,32,33,28.7,35,29,29]
ice_creams = [100, 120, 130, 75, 125, 77, 83]

plt.scatter(temp, ice_creams)
plt.title("Temperature vs IceCream sold")
plt.xlabel("Temp")
plt.ylabel("Ice Cream sold")
```

```python
#histogram chart - groups for us numbers in rangess -  bins - the higher the bar is the more
#data exists for the range/bin
numbers = [1,2,3,3.5,4,4.5,4,6.3,6.6,6.7,7,8,9,9.2,9.3,9.5]

plt.hist(numbers,bins=5)
plt.title("Histogram of numbers")
plt.xlabel("numbers")
plt.ylabel("Frequency")
```

```python
#bar/column chart/plot

labels = ["Javascript", "Java", "Python", "R"]
usage = [62.8, 43.5, 37.2, 25]

#plotting the bar itself
plt.bar(labels, usage)
plt.title("Programming languages usage")
plt.xlabel("Programming languages")
plt.ylabel("Usage")
```

```python
planguages = ["Javascript", "Java", "C++", "Python", "R"]#Qualitative data
usage = [82.5, 79.2, 60.3, 100, 45.9]#Quantitative data, e.g in terms of millions

#get horizontal bar chart
plt.barh(planguages, usage)
plt.title("Programming languages usage")
plt.ylabel("Programming languages")
plt.xlabel("Usage")
```

```python
#a histogram is like a bar chart - it takes in a continous data and bins it!
#it groups your continous data into ranges

#the higher a bar is, the higher the concentration of occurrence
Student_Scores = [20,33,26,28,45,60,61,62,60,63,67,80,85,75,85,81,55,56,64,58,95,99]

plt.hist(Student_Scores, bins=5)
plt.title("Student score groupings")
plt.xlabel("Student Scores")
plt.ylabel("Count")
```
