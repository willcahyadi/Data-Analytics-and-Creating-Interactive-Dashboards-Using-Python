## Data pre-processing and data manipulation

```python
#import some libraries

import pandas as pd
import numpy as np
```

```python
url = "https://bit.ly/2Qk25Xx"
df = pd.read_csv(url)
df.head(10)
```

```python
df.info()
```

```python
# first approah to cleaning up our data is to remove the rows with missing values
df = df.dropna()
```

```python
df.info()
```

```python
len(df["ARREST_KEY"].unique())
```

```python
df = df.set_index("ARREST_KEY")
df
```

```python
df.info()
```

```python
# we know some columns may not be useful for us, we can drop them too
dropcols = ["LAW_CODE", "ARREST_BORO", "ARREST_PRECINCT", "X_COORD_CD", "Y_COORD_CD", "Latitude",
           "Longitude"]
df = df.drop(columns=dropcols)
df
```

### Create new columns
inside LAW_CAT_CD
> F is for felony - 36 months of jail time
>
> M is for misdemeanors - 12 months of jail time
>
> V is for violation - 1 month of jail time

```python
#numpy - an explainer
np.where(1>2, "apple", 
         np.where(3>2, "watermelon", "pear") #this is in the else clause
        )
```

```python
df["JAIL_TIME"] = np.where(df["LAW_CAT_CD"] == "F", 36, 
         np.where(df["LAW_CAT_CD"] == "M", 12, 1)
        )
```

```python
df.head()
```

#### this idea of creating new columns has a name - Feature Engineering!

### Dropna is a really destructive technique, we can potentially lose alot of data when we use drop na!

```python
#create a small data
ndata = pd.Series([np.nan, np.nan, 3, None, 2])
ndata
```

```python
#method 1: is to fill using a specific number
#typical way is to fillm NaN with the number 0, we can use a meaningful number such as the
#mean, or the median
nmean = ndata.mean()
ndata.fillna(nmean) #fill with the mean
```

```python
ndata.fillna(0)
```

```python
ndata
```

```python
#method 2: forward fill
ndata.fillna(method="ffill")
```
