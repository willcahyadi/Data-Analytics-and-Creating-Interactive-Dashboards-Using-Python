```python
#dataframes - its multi dimensional data structure
data = {'country': ['Singapore', 'France', 'Germany'],
        'population': [5.5, 64, 82],'area': [31000, 672000, 360000],
        'capital': ['Singapore', 'Paris', 'Berlin']}

#convert into a dataframe
countrydf = pd.DataFrame(data)
countrydf
```

## INDEXING AND SLICING DATA

#### Indexing - Being able to access different parts of the data using rows and columns index
#### Slicing - Being able to extract a subset of data using index of rows/columns

```python
countrydf
```

```python
# i want the areas and population of singapore and france
#loc - row locations and column column locations
countrydf.loc['Singapore':"France", 'population':'area']#row, col
```

```python
#sau we want populatioon and capital
popcap = ['population','capital']
countrydf.loc['Singapore':"France", popcap]#row, col
```

```python
countrydf.loc['Singapore':"France", popcap].describe()
```

## Filtering - We want to subset our data - where area of the country is more than 100000

```python
mask = countrydf["area"] > 100000
countrydf[mask] # this is filtering
```

### Using index - iloc to manipulate our data

```python
#dataframes - its multi dimensional data structure
data = {'country': ['Singapore', 'France', 'Germany'],
        'population': [5.5, 64, 82],'area': [31000, 672000, 360000],
        'capital': ['Singapore', 'Paris', 'Berlin']}

#convert into a dataframe
countrydf = pd.DataFrame(data)
countrydf
```

```python
## simulate this  - countrydf.loc['Singapore':"France", 'population':'area']#row, col
countrydf.iloc[0:2, 1:3] #row, col
```

#### Say we got the area of singapore wrong, we can change the area using the following:

```python
countrydf.iloc[0, 2] = 40000
countrydf
```

```python
## when we created the population density pandas series we can also similarly create a column
#From the country perspective, we want find the population density of every country
#population / area
countrydf['popdensity'] = countrydf["population"] / countrydf["area"]
countrydf
```

```python
#rename the column
countrydf = countrydf.rename(columns={'popdensity':'population_density'})
countrydf
```
